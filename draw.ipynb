```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar la figura para usar coordenadas polares
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, projection='polar')

# Parmetros del rosetn
n = 6  # Nmero de "ptalos" (puedes ajustarlo)
theta = np.linspace(0, 2 * np.pi, 1000)  # ngulos de 0 a 2

# Ecuacin para un rosetn (puedes experimentar con esta frmula)
r = np.cos(n * theta)

# Graficar el rosetn
ax.plot(theta, r, color='red', linewidth=2)

# Personalizar el estilo medieval
ax.set_rticks([])  # Eliminar las lneas radiales
ax.set_yticklabels([])  # Eliminar etiquetas radiales
ax.set_xticks(np.linspace(0, 2 * np.pi, n, endpoint=False))  # Divisiones angulares
ax.set_xticklabels([])  # Eliminar etiquetas angulares
ax.set_facecolor('#f5f5f5')  # Fondo claro para un toque medieval
ax.grid(False)  # Sin rejilla para un diseo ms limpio

# Ttulo (opcional)
plt.title("Rosetn Medieval", fontsize=16, pad=20)

# Mostrar la grfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar la figura para coordenadas polares
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, projection='polar')

# Parmetros de la mandala
theta = np.linspace(0, 2 * np.pi, 1000)  # ngulos de 0 a 2

# Capas de la mandala (combinacin de patrones)
r1 = np.sin(8 * theta)  # Patrn base con 8 ptalos
r2 = 0.5 * np.cos(16 * theta) + 1  # Capa secundaria ms detallada
r3 = 0.3 * np.sin(32 * theta) + 1.2  # Detalles finos

# Graficar las capas
ax.plot(theta, r1, color='gold', linewidth=1.5, alpha=0.8)
ax.plot(theta, r2, color='red', linewidth=1, alpha=0.7)
ax.plot(theta, r3, color='blue', linewidth=0.8, alpha=0.6)

# Personalizar el estilo budista
ax.set_rticks([])  # Sin lneas radiales
ax.set_yticklabels([])  # Sin etiquetas radiales
ax.set_xticks([])  # Sin divisiones angulares
ax.set_facecolor('#f0f0f0')  # Fondo claro
ax.grid(False)  # Sin rejilla para un diseo puro

# Rellenar reas para ms profundidad (opcional)
ax.fill(theta, r1, 'gold', alpha=0.3)
ax.fill(theta, r2, 'red', alpha=0.2)

# Ttulo (opcional)
plt.title("Mandala Budista", fontsize=18, pad=20)

# Mostrar la mandala
plt.show()
```
